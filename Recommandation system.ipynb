```python
#converting the dataset to matrix format using pivot table
```

```python
book_df=book.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
book_df
```

```python
#changing the index with userid
```

```python
book_df.index =b
```

```python
book_df
```

```python
#filling NaN with 0 for computation
```

```python
book_df.fillna(0,inplace=True)
```

```python
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_sim = 1 - pairwise_distances(book_df.values,metric='cosine')
```

```python
book_sim
```

```python
#Store the results in a dataframe
book_sim_df = pd.DataFrame(book_sim)
book_sim_df
```

```python
#filling the diagonal values 0 because all diagonal values will be 1 for same user id
```

```python
np.fill_diagonal(book_sim,0)
book_sim_df
```

```python
#Set the index and column names to user ids 
book_sim_df.index = b
book_sim_df.columns = b
book_sim_df
```

```python
#These are the most similar users in our dataset with same choices
```

```python
similar=book_sim_df.idxmax(axis=1)
similar
```
